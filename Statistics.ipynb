```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(
    {
        "student" : [20,30,40,50,80,90,155,500],
        'age' : [23,24,26,45,23,23,20,22]
    }
)
df
```

```python
vari = np.var(df["student"])
print(vari)
stder = np.std(df["student"])
print(stder)
me = np.mean(df['student'])
print(me)
```

```python
plt.hist(df['student'])

plt.axvline(me, color='red', linestyle='dashed', linewidth=2)
```

```python
percent = np.percentile(df['age'],80)
print(percent)
```

```python
# quartile
q1 = np.percentile(df['age'], 25)
q2 = np.percentile(df['age'], 50)
q3 = np.percentile(df['age'], 75)
q4 = np.percentile(df['age'], 100)

print(q1, q2, q3, q4)
```

```python
hourse = pd.DataFrame(
    {
        "H" : [200,300,434,256,654,700,355,700,1200]
    }
)
hourse.sort_values(by="H",inplace=True)
print(hourse)
print("mead", np.mean(hourse["H"]))
print("varience" , np.var(hourse["H"]))
print("Stander dev", np.std(hourse["H"]))
q0 = np.percentile(hourse['H'],0)
q1 = np.percentile(hourse["H"],25)
q2 = np.percentile(hourse["H"],50)
q3 = np.percentile(hourse["H"],75)
q4 = np.percentile(hourse["H"],100)
print("percentile and Qu..." ,q0, q1, q2 ,q3 ,q4 )
```

```python
from matplotlib import lines
plt.figure(figsize=(12,6))
plt.hist(hourse["H"])
plt.axvline(q0,color="black", linestyle="dashed", linewidth=2,label="Q1")
plt.axvline(q1,color="red", linestyle="dashed", linewidth=2,label="Q1")
plt.axvline(q2,color="green", linestyle="dashed", linewidth=2,label="Q2")
plt.axvline(q3,color="yellow", linestyle="dashed", linewidth=2,label="Q3")
plt.axvline(q4,color="pink", linestyle="dashed", linewidth=2,label="Q4")

plt.title("House Data")
plt.xlabel("House Price")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis="y", linestyle="--",alpha=0.4)
plt.show()

```

```python
# IQR = q4 - q3  mane full datar 75% - 25% data sub korle IQR ber hoy

iqr =  q3 - q1
print(iqr)
```

```python
# to detect outlier standard rule(Tukey's method) 1.5 kom besi kora jay kintu 1.5 standard
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr
print(upper, lower)
```
